```python
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
```

```python
PI = np.pi

x = np.linspace(0, 2*PI, 50)
fig, ax = plt.subplots()
y1 = ax.plot(x, np.sin(x), '--', dashes = [10, 5, 100, 5], label='Sin (x)')
y2 = ax.plot(x, np.cos(x), '-', dashes = [30, 5, 10, 5], label='Cos (x)')

ax.set_xlim((0, 2*PI))
ax.set_ylim((-1.5, 1.5))

ax.set_xticks([0, PI, 2*PI])
ax.set_xticklabels(['0', '$\pi$', '2$\pi$'])
ax.set_yticks([-1, 0, 1])
ax.legend(loc='lower right')
plt.show()
```
